```python
dfshow = pd.pivot_table(
    df_formulacao1.query("amount_of_end_products == 5"),
    index=col_keys3,
    aggfunc={
        "total_backlogged_end_products": np.mean,
        "ingredients_uc": np.mean,
        "end_product_uc": np.mean,
    },
).reset_index().rename(
    columns={
        "capacity_multiplier": "capacidade produtos finais",
        "coef_cap": "multiplicador cap ingredients",
    }
)
dfshow = dfshow[dfshow.apply(lambda row: not isinstance(row["capacidade produtos finais"], str), axis=1)]
#show(dfshow)
```

```python
pd.pivot_table(
    df_formulacao1.query("amount_of_end_products == 5 and type_cap_ingredients == 'W'"),
    index=col_keys4,
    aggfunc={
        "total_backlogged_end_products": np.mean,
        "ingredients_uc": np.mean,
        "end_product_uc": np.mean,
    },
).reset_index().rename(
    columns={
        "capacity_multiplier": "capacidade produtos finais",
        "coef_cap": "multiplicador cap ingredients",
    }
)
```

```python
dfaux = pd.pivot_table(
    df_formulacao1.query("type_cap_ingredients != 'W'"),
    index=col_keys3,
    aggfunc={
        "total_backlogged_end_products": np.mean,
        "ingredients_uc": np.mean,
        "end_product_uc": np.mean,
    },
).reset_index().rename(
    columns={
        "capacity_multiplier": "capacidade produtos finais",
        "coef_cap": "multiplicador cap ingredients",
    }
)
dfaux = dfaux[dfaux.apply(lambda row: not isinstance(row["capacidade produtos finais"], str), axis=1)]
#show(dfshow)



fig = px.density_heatmap(
    dfaux,
    x="capacidade produtos finais",
    y="multiplicador cap ingredients",
    z="total_backlogged_end_products",
    text_auto=True,
    #marginal_x="histogram", marginal_y="histogram",
    histfunc="avg",
    facet_row="amount_of_end_products"
)

fig.update_layout(
    margin=dict(l=20, r=20, t=20, b=20),
    width=800,
    height=680
)

fig.show()
```

```python
fig = px.density_heatmap(
    dfaux,
    x="capacidade produtos finais",
    y="multiplicador cap ingredients",
    z="ingredients_uc",
    text_auto=True,
    #marginal_x="histogram", marginal_y="histogram",
    histfunc="avg",
    facet_row="amount_of_end_products"
)

fig.update_layout(
    margin=dict(l=20, r=20, t=20, b=20),
    width=800,
    height=680
)

fig.show()
```

```python
fig = px.density_heatmap(
    dfaux,
    x="capacidade produtos finais",
    y="multiplicador cap ingredients",
    z="end_product_uc",
    text_auto=True,
    #marginal_x="histogram", marginal_y="histogram",
    histfunc="avg",
    facet_row="amount_of_end_products"
)

fig.update_layout(
    margin=dict(l=20, r=20, t=20, b=20),
    width=800,
    height=680
)

fig.show()
```

```python
dfaux["raiz_pXi"] = (dfaux["end_product_uc"]*dfaux["ingredients_uc"])**(0.5)
fig = px.density_heatmap(
    dfaux,
    x="capacidade produtos finais",
    y="multiplicador cap ingredients",
    z="raiz_pXi",
    text_auto=True,
    #marginal_x="histogram", marginal_y="histogram",
    histfunc="avg",
    facet_row="amount_of_end_products"
)

fig.update_layout(
    margin=dict(l=20, r=20, t=20, b=20),
    width=800,
    height=680
)

fig.show()
```
