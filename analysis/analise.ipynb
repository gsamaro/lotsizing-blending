```python
col_keys2 = [
    "Instance",
    "capacity_multiplier",
    "type_cap_ingredients",
    "coef_cap",
    "amount_of_end_products",
]
col_keys3 = ["capacity_multiplier", "coef_cap", "amount_of_end_products"]
col_keys4 = col_keys3 + ["type_cap_ingredients"]
```

```python
pd.pivot_table(
    df_formulacao1,
    index=col_keys3,
    aggfunc={"ingredients_uc": np.mean, "end_product_uc": np.mean},
).reset_index().rename(
    columns={
        "capacity_multiplier": "capacidade produtos finais",
        "coef_cap": "multiplicador cap ingredients",
    }
)
```

```python
pd.pivot_table(
    df_formulacao1,
    index=col_keys3,
    columns=["status"],
    aggfunc={"status": "count", "Gap": np.mean},
).reset_index().rename(
    columns={
        "capacity_multiplier": "capacidade produtos finais",
        "coef_cap": "multiplicador cap ingredients",
    }
)
```

```python
pd.pivot_table(
    df_formulacao1, index=col_keys3, aggfunc={"Gap": np.mean}
).reset_index().rename(
    columns={
        "capacity_multiplier": "capacidade produtos finais",
        "coef_cap": "multiplicador cap ingredients",
    }
)
```

```python
pd.pivot_table(
    df_formulacao1.query("amount_of_end_products == 5"),
    index=col_keys3,
    aggfunc={
        "total_backlogged_end_products": np.mean,
        "ingredients_uc": np.mean,
        "end_product_uc": np.mean,
    },
).reset_index().rename(
    columns={
        "capacity_multiplier": "capacidade produtos finais",
        "coef_cap": "multiplicador cap ingredients",
    }
)
```

```python
pd.pivot_table(
    df_formulacao1.query("amount_of_end_products == 5 and type_cap_ingredients == 'W'"),
    index=col_keys4,
    aggfunc={
        "total_backlogged_end_products": np.mean,
        "ingredients_uc": np.mean,
        "end_product_uc": np.mean,
    },
).reset_index().rename(
    columns={
        "capacity_multiplier": "capacidade produtos finais",
        "coef_cap": "multiplicador cap ingredients",
    }
)
```
